## 식별자

- 작성 방법

  - 첫번째 자리에 숫자가 올 수 없음 ex) `0name` (x)

  - 내장함수 or 모듈 등의 이름으로 만들면 안됨
    - 예약어, 내장함수, 모듈 등의 이름으로 변수 생성시, 기존의 기능이 덮여쓰여짐 


# 변수 및 기본 자료형

## 자료형
### 정수(integer)형
- 정수 값
- -2, 1, 0 ,10 ...
- `type(10)` >>> class 'int'

### 문자열(string)형
- 문자나 문자열을 `""`나 `''`로 감싸서 표현
- 'A', 'a', 'age', '나이' ...
- `type('가')` >>> class 'str'

### 실수(float)형
- 소수점 이하의 정밀한 값 표현
- 2.1, 1.12543875 ...
- `type('2.1')` >>> class 'float'

### 부울(bool)형
- 참 or 거짓, true or false
- `a = (10==10)`
- `type(a)` >>> class 'bool'

## 변수
- `=`를 통해 할당함
- `type()`으로 자료형 확인이 가능
- 특정 변수의 메모리 주소를 확인하기 위해서는 `id()`를 확인
- 같은 값을 동시에 할당할 수 있음 
  - `a = b = 10`
- 다른 값을 동시에 할당할 수 있음
- `a, b = 1, 2`

````python
# stack_push(스택 변수, 접시 가격) 함수 정의
def stack_push(stack, price): # 함수 선언(정의)
    stack.append(price)
    
# stack_pop(스택 변수) 함수 정의
# 스택변수 길이가 0이면 "없음" 반환, 스택변수 마지막 데이터 삭제(pop)하여 반환
def stack_pop(stack):
    if len(stack) == 0:
        return '없음'
    return stack.pop()

print("[초밥 접시 계산하기]")

# 초밥 접시 리스트
stack = [] 

# 총 가격
totalPrice = 0

# 스택 자료구조 리스트로 초기화
while True:
    price = input('접시가격 : ')
    price = int(price)
    if price == 0:
        break
    stack_push(stack, price) #함수 호출

print('쌓인 접시 : ', stack)
print()

while True:
    pop_data = stack_pop(stack)
    if pop_data == '없음':
        break
    totalPrice += int(pop_data)
    print('계산접시 : ', pop_data)
    print('남은접시 : ', stack)

print('계산 완료')

print('총 가격 : ', totalPrice)
````
