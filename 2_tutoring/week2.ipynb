# 도전과제 2번
- 참고로 `.grid()`와 `.pack()`은 함께 쓸 수 없음 -> 둘중 하나만 써야함
    - Grid : 격자 배치 관리자(grid geometry manager)
    테이블 형태의 배치, grid() 메소드 이용
    - Pack : 압축 배치 관리자(pack geometry manager) 
    위젯들을 부모 위젯 안에 압축 
    자식 위젯들을 사각형 블록으로 간주하여 프레임 안에 배치
    pack()메소드 사용
    - Place : 절대 배치 관리자(place geometry manager) 
    주어진 위치에 위젯을 배치, place() 메소드 사용

```python
from tkinter import *

def result_sum():
    num1 = int(n1.get())
    num2 = int(n2.get())
    res = num1+num2
    result.config(text=res)

def result_sub():
    num1 = int(n1.get())
    num2 = int(n2.get())
    res = num1-num2
    result.config(text=res)

def result_mul():
    num1 = int(n1.get())
    num2 = int(n2.get())
    res = num1*num2
    result.config(text=res)

def result_div():
    num1 = int(n1.get())
    num2 = int(n2.get())
    res = num1/num2
    result.config(text=res)


s=Tk()
s.title('사칙연산 계산기')

label1=Label(s, text='숫자1 입력')
label1.grid(row=1, column=0, columnspan=2)
n1=StringVar()
entry_n1 = Entry(s, textvariable=n1).grid(row=1, column=2,columnspan=2)

label2=Label(s, text='숫자2 입력')
label2.grid(row=2,column=0, columnspan=2)
n2=StringVar()
entry_n2 = Entry(s, textvariable=n2).grid(row=2, column=2,columnspan=2)


btn1 = Button(s, width=10, text="+",command=result_sum)
btn1.grid(row=3, column=0)
btn2 = Button(s, width=10, text="-",command=result_sub)
btn2.grid(row=3, column=1)
btn3 = Button(s, width=10, text="*",command=result_mul)
btn3.grid(row=3, column=2)
btn4 = Button(s, width=10, text="/",command=result_div)
btn4.grid(row=3, column=3)

# 결과
result=Label(s,text='',font=('맑은 고딕',12),bg='salmon',width=30, anchor="w")
result.grid(row=4, column=0, columnspan=4)
s.mainloop()
```
